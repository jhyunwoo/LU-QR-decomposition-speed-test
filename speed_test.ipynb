```python
import scipy
import scipy.linalg
import numpy as np
import time
```

```python
start_time = time.time()
cal_times_qr = np.array([])

for i in range(1, 1000):
    cal_time = time.time()
    np.random.seed(i)
    A = np.random.rand(i, i)
    Q, R = scipy.linalg.qr(A)
    print(i, "claculate time is", time.time() - cal_time)
    cal_times_qr = np.append(cal_times_qr, np.array([time.time() - cal_time]))

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()
cal_times_lu = np.array([])

for i in range(1, 1000):
    cal_time = time.time()
    np.random.seed(i)
    A = np.random.rand(i, i)
    P, L, U = scipy.linalg.lu(A)
    print(i, "claculate time is", time.time() - cal_time)
    cal_times_lu = np.append(cal_times_lu, np.array([time.time() - cal_time]))

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,1000)
y_qr = cal_times_qr
plt.plot(x, y_qr)
```

```python
x = np.arange(1,1000)
y_lu = cal_times_lu
plt.plot(x, y_lu)
```
