```python
import cupy
import cupy.linalg
import cupyx.scipy
import cupyx.scipy.linalg
import cupy as cp
import time
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
lu_cal_time = np.array([])
qr_cal_time = np.array([])

for j in range(1,1000):

  # LU Decomposition
  start_time = time.time()

  for i in range(1, 1000):
    cp.random.seed(i)
    A = cp.random.rand(j, j)
    P, L, U = cupyx.scipy.linalg.lu(A)

  print("LU Decomposition", j, "x",j, "array ---", time.time() - start_time, "sec")
  lu_cal_time = cp.append(lu_cal_time, cp.array([time.time() - start_time]))


  # QR Decomposition
  start_time = time.time()

  for i in range(1, 1000):
      cp.random.seed(i)
      A = cp.random.rand(j, j)
      Q, R = cupy.linalg.qr(A)

  print("QR Decomposition", j, "x",j,"array ---", time.time() - start_time, "sec")
  qr_cal_time = cp.append(qr_cal_time, cp.array([time.time() - start_time]))
```

```python
x = np.arange(1,1000)
y_lu = lu_cal_time.get()
y_qr = qr_cal_time.get()
plt.figure(figsize=(10,7))
plt.title("LU Decomposition & QR Decomposition Speed Test")
plt.plot(x, y_lu, x, y_qr, 'r-')
```

```python
for i in range(1, 999):
  if lu_cal_time[i]<qr_cal_time[i]:
    print(i, "lu is better")
```
